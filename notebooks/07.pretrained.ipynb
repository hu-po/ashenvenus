```python
import logging

import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision.models import convnext_tiny, ConvNeXt_Tiny_Weights
```

```python
class PreTrainNet(nn.Module):
    def __init__(
        self,
        slice_depth: int = 65,
        # pretrained_weights_filepath = '/kaggle/input/convnextimagenet/convnext_tiny-983f1562.pth',
        pretrained_weights_filepath = '/home/tren/dev/ashenvenus/notebooks/convnext_tiny-983f1562.pth',

    ):
        super().__init__()
        self.conv = nn.Conv2d(slice_depth, 3, 3)
        # Use pretrained model
        self.model = convnext_tiny()
        self.model.load_state_dict(torch.load(pretrained_weights_filepath))
        self.fc = nn.LazyLinear(1)

    def forward(self, x):
        x = self.conv(x)
        x = self.model(x)
        x = self.fc(x)
        return x
```

```python
INPUT_SHAPE = (32, 65, 224, 224)
OUTPUT_SHAPE = (32, 1)

model = PreTrainNet()

# Fake data to test the model
x = torch.randn(INPUT_SHAPE, dtype=torch.float32)
y = model(x)
```
